```python
from top_ori import read_trees as read_trees_ori
from top import read_trees, diff, tree_level
from utils import Statistics, merge_statistics
```

```python
seeds = "123 456 789 555 666".split()
methods = "seq2seq.base seq2seq.large span.base".split()

gold_trees= read_trees("top/gold.parse")
for m in methods:
    for s in seeds:
        if m.startswith("span"):
            pred_trees = read_trees_ori(f"top/{m}{s}.parse")
        else:
            pred_trees = read_trees(f"top/{m}{s}.parse")
        acc = 0
        invalid = 0
        for a, b in zip(gold_trees, pred_trees):
            if b["status"] == "invalid":
                invalid +=1
                
            if b["status"] == "valid" and not diff(a["root"], b["root"]):
                acc += 1
        acc = acc / len(gold_trees)
        print (f"{m}{s}:{acc*100:.2f}")
        print (f"invalid: {invalid}, total: {len(pred_trees)}, rate: {invalid/len(pred_trees)*100:.2f}")
```

```python
hold_all = {}
stats = []
m1 = "span.base"
m2 = "seq2seq.base"
for s1 in seeds:
    for s2 in seeds:
        print (f"{m1}{s1}=>{m2}{s2}")
        res = tree_level("top/gold.parse", read_trees_ori("top/"+m1+s1+".parse"), "top/"+m2+s2+".parse")
        stat = merge_statistics(res)
        stats.append(stat)
print ("micro average")
stat = merge_statistics(stats)
hold_all[f"{m1}.{m2}"] = stat
stat.report(["atom", "whole"])
```

```python
for m in methods:
    stats = []
    for s1 in seeds:
        for s2 in seeds:
            if s1!=s2:
                print (f"{m}:{s1}=>{s2}")
                if m.startswith("span"):
                    res = tree_level("top/gold.parse", read_trees_ori("top/"+m+s1+".parse"), read_trees_ori("top/"+m+s2+".parse"))
                else:
                    res = tree_level("top/gold.parse", "top/"+m+s1+".parse", "top/"+m+s2+".parse")
                stat = merge_statistics(res)
                stats.append(stat)

    print ("micro average")
    stat = merge_statistics(stats)
    hold_all[f"{m}"] = stat
    stat.report(["atom", "whole"])
```

```python
which = 'span.base.seq2seq.base'# "span.base"

def breakdown(x, prefix, max_rows=None):
    keys = [k for k in x.keys if k.startswith(prefix)]
    print (f"{prefix.split('#')[0]}#NRF#PFR#old ACC#new ACC#NFI#Count")
    for k in sorted(keys, key=lambda k:-x[k].nfr)[:max_rows]:
        xxx = k.split('#')[-1]
        x[k].report_table(f"{xxx}")
    print ("="*55)

for prefix in ["span#", "tspan#", "size#", "depth#", "IN#", "SL#", "linearization#"]:
    if prefix == "linearization#":
        breakdown(hold_all[which], prefix, 10)
    else:
        breakdown(hold_all[which], prefix)
```
